```python
#importing librairies
import numpy as np #math tools
import matplotlib.pyplot as plt
import pandas as pd #import and manage datasets

#importing the dataset
dataset_train=pd.read_csv('../titanic_data/train.csv')
X_train=dataset_train.iloc[:,[2,4,5,6,7,9,11]].values
X_train=pd.DataFrame(X_train) #data.iloc[row selection,column selection] -1 retire le dernier
y_train=dataset_train.iloc[:,1].values
dataset_test=pd.read_csv('../titanic_data/test.csv')
X_test=dataset_test.iloc[:,[1,3,4,5,6,8,10]].values
X_test=pd.DataFrame(X_test)

#OneHotEncoding Dataframes
X_train=pd.concat([X_train,pd.get_dummies(X_train[0])],axis=1, ignore_index=True)
X_train=X_train.drop([0,len(X_train.T)-1],axis=1)
X_train=pd.concat([X_train,pd.get_dummies(X_train[6])],axis=1, ignore_index=True)
X_train=X_train.drop([5,len(X_train.T)-1],axis=1)

X_test=pd.concat([X_test,pd.get_dummies(X_test[0])],axis=1, ignore_index=True)
X_test=X_test.drop([0,len(X_test.T)-1],axis=1)
X_test=pd.concat([X_test,pd.get_dummies(X_test[6])],axis=1, ignore_index=True)
X_test=X_test.drop([5,len(X_test.T)-1],axis=1)

#taking care of missing data 

def mean_imputer (DataFrame, column_index):
    total=0
    cpt=0
    mean=0
    for i in range (len(DataFrame[column_index])):
        if (np.isnan(DataFrame[column_index][i])==False):
            total+=DataFrame[column_index][i]
            cpt+=1
    mean=total/cpt
    for i in range (len(DataFrame[column_index])):
        if (np.isnan(DataFrame[column_index][i])):
           DataFrame[column_index][i]=mean

"""find where are 'nan':
    for i in range (len(X_train.T)+1):
    for j in range (len(X_train)):
        if (np.isnan(X_train[i][j])):
            print (i)
            print (j)"""

mean_imputer (X_train, 1)
mean_imputer (X_test, 1)
mean_imputer (X_test, 4)

#LabelEncoding man/woman
from sklearn.preprocessing import LabelEncoder
labelencoder_X=LabelEncoder();
X_train[0][:]= labelencoder_X.fit_transform(X_train[0][:])
X_test[0][:]= labelencoder_X.transform(X_test[0][:])

#Scaling data
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
X_train=Scaler.fit_transform(X_train)
X_test=Scaler.transform(X_test)

# Applying PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 5)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_

from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X_train, y_train, test_size=0.2)

# Fitting SVM to the Training set
from sklearn.svm import SVC
classifier = SVC(gamma=0.3, kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)

pred_y = classifier.predict(test_X)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_y, pred_y)

#Evaluating accuracy
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(test_y, pred_y)



"""# Applying Grid Search to find the best model and the best parameters
from sklearn.model_selection import GridSearchCV
parameters = [{'C':[0.9,1,1.1]},
              {'kernel': ['rbf','linear','poly','sigmoid'], 'gamma': [0.25, 0.3, 0.35]}]
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10,)
grid_search = grid_search.fit(X_train, y_train)
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_"""


# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, scoring = 'accuracy', X = X_train, y = y_train, cv = 10)
accuracies.mean()
accuracies.std()

# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = pd.DataFrame(y_pred)

prediction=pd.concat([dataset_test['PassengerId'],y_pred],axis=1, ignore_index=True)

prediction = prediction.to_csv('prediction.csv', header=['PassengerId','Survived'], index=False)
```
